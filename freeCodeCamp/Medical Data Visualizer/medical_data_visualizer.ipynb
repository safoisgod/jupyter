```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
import numpy as np
```

```python
pd.set_option("display.max_rows",20)
pd.set_option("display.max_columns",None)
```

```python
df=pd.read_csv("medical_examination.csv")
df
```

## Add Overweight Column

### create BMI column

```python
df["height"]=df["height"]/100
df["BMI"]=df["weight"]*(df["height"]**2)
df
```

### create and assign overweight values using lambda

```python
df["overweight"] = df["BMI"].apply(lambda bmi: 1 if bmi > 25 else 0)
df
```

### normalize Cholestrol

```python
df["cholesterol"]=df["cholesterol"].apply(lambda cholesterol : 1 if cholesterol >1 else 0)
df
```

### normalize Glucose

```python
df["gluc"]=df.gluc.apply(lambda gluc: 1 if gluc>1 else 0)
df
```

## Draw Category Plot

Create Dataframe for the Categorical Plot for "cholestrol", "gluc", "smoke", "alco", "cardio" and "overweight"

use pd.melt
>pd .melt converts the data into a long format
>>it selects on column as the "ID column(id_vars)", then other columns can be  selected as variables, where values in those columns are shown in the values column

```python
df_cat = pd.melt(df, id_vars=["cardio"], value_vars=["cholesterol","gluc","smoke","alco","cardio","overweight"])
df_cat
```

use groupby to categorise the cardio, variable and value, based on the size(rows)

```python
df_cat = df_cat.groupby(['cardio', 'variable', 'value']).size().reset_index(name="total")
df_cat
```

use seaborn.catplot(kind="bar") to visualize data
>col, helps craete subplots based on values of the column specified
>>hue, helps to groupby the column specified

```python
sns.catplot(data=df_cat, kind="bar", y="total",x="variable", col="cardio", hue="value")
```

## Clean Data

based on the data given, the following should not be in the data
- diastolic pressure is higher than systolic
- height is less than the 2.5th percentile
- height is more than the 97.5th percentile
- weight is less than the 2.5th percentile
- weight is more than the 97.5th percentile

#### therefore the opposite will be filtered for and re-assigned to the dataframe

```python
df=df[(df['ap_lo'] <= df['ap_hi'])&
    (df['height'] >= df['height'].quantile(0.025))&
    (df['height'] <= df['height'].quantile(0.975))&
    (df['weight'] >= df['weight'].quantile(0.025))&
    (df['weight'] <= df['weight'].quantile(0.975))]
df
```

#### Create correlation matrix for the data

```python
corr = df.corr()
corr
```

#### mask of upper triangle 

```python
mask=np.triu(np.ones_like(corr))
mask
```

#### set up matplotlib figure
#### create heatmap using seaborn
- set data to correlation, annotation(caliberation, text) to True, formatting to .1f, square to true to keep all cells in square shape, and the mask to the mask variable

```python
fig, ax = plt.pyplot.subplots(figsize=(16,9))
sns.heatmap(corr, annot=True, fmt="0.1f", square=True, mask=mask)
```
